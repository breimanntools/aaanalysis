# `AnnotationPreprocessor.register_feature()`

`register_feature` explicitly registers (or overrides) an open-vocabulary **Functional sites** key before ingest — giving it a `subcategory` label and an optional custom `normalization` callable (default: identity, values must already lie in `[0, 1]`). Use it when you want a predictor channel named and normalized your way rather than auto-registered.

```python
import warnings
import numpy as np
import pandas as pd
import aaanalysis as aa
import aaanalysis.utils as ut
aa.options['verbose'] = False
warnings.filterwarnings('ignore')

annp = aa.AnnotationPreprocessor(verbose=False)
df_seq = pd.DataFrame({'entry': ['AF_TINY'],
                       'sequence': ['ACDEFGHIKLMNPQRSTVWYACDEFGHIKL']})

annp.register_feature(key='kinase_site',
                    subcategory='Kinase site (predictor)')
print('registered:', 'kinase_site' in annp._registry)
print(annp._registry['kinase_site'])
```

```python
# Further parameter: attach a custom ``normalization`` callable (default: identity,
# so inputs must already lie in [0, 1]). Re-registering the same key overrides it.
annp.register_feature(key='kinase_site',
                      subcategory='Kinase site (predictor)',
                      normalization=lambda x: np.clip(x, 0.0, 1.0))
print('normalization registered for:', annp._registry['kinase_site']['dim_names'])
```
