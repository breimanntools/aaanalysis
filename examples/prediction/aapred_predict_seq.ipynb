``AAPred().predict_seq()`` scores one value per protein directly from a ``df_seq``. We bind ``df_feat`` to the model so the feature matrix is computed internally (see [Breimann25]_):

```python
import aaanalysis as aa
aa.options["verbose"] = False  # Disable verbosity

df_seq = aa.load_dataset(name="DOM_GSEC")
labels = df_seq["label"].to_list()
df_feat = aa.load_features(name="DOM_GSEC").head(20)

# Training feature matrix
sf = aa.SequenceFeature()
X = sf.feature_matrix(features=df_feat, df_parts=sf.get_df_parts(df_seq=df_seq))

# Bind df_feat so raw sequences are featurized internally by the predict_* methods
aapred = aa.AAPred(df_feat=df_feat, random_state=42)
aapred.fit(X, labels)
```

Each protein gets a positive-class score and its across-model standard deviation:

```python
df_pred = aapred.predict_seq(df_seq.head(10))
aa.display_df(df_pred, n_rows=10, show_shape=True)
```

**Further parameters.** ``list_parts`` selects which sequence parts are built for featurization (defaults to the standard CPP parts):

```python
df_pred = aapred.predict_seq(df_seq.head(10), list_parts=["tmd", "jmd_n_tmd_n", "tmd_c_jmd_c"])
aa.display_df(df_pred, n_rows=10, show_shape=True)
```
