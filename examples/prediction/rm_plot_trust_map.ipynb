``ReliabilityModelPlot().trust_map()`` plots prediction score vs. OOD score, colored by
the ``reliable`` flag — a confident point above the in-domain boundary is an untrustworthy
extrapolation.

```python
import aaanalysis as aa
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
aa.options["verbose"] = False
aa.plot_settings()
X, labels = make_classification(n_samples=160, n_features=10, n_informative=6, random_state=42)
X_train, labels_train, X_new = X[:120], labels[:120], X[120:]
rm = aa.ReliabilityModel(random_state=42).fit(X=X_train, labels=labels_train)
df_rel = rm.predict(X=X_new)
df_eval = rm.eval(X=X_new, labels=labels[120:])
```

```python
aa.ReliabilityModelPlot().trust_map(df_rel=df_rel, figsize=(5.5, 5),
                                    title="Trust map", ax=None)
plt.tight_layout()
plt.show()
```
