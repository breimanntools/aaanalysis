To demonstrate ``AAPredPlot().clustermap()``, we cluster samples by explanation similarity — the correlation of their per-sample importance vectors (e.g. SHAP values). Here we use a small synthetic importance matrix; in practice these come from :class:`ShapModel` (``pro``):

```python
import numpy as np
import aaanalysis as aa
aa.options["verbose"] = False

rng = np.random.RandomState(0)
# Two groups of samples with distinct explanation patterns.
block_a = rng.normal(0.6, 0.2, size=(6, 20))
block_b = rng.normal(-0.6, 0.2, size=(6, 20))
data = np.vstack([block_a, block_b])
labels = np.array([1] * 6 + [0] * 6)
names = [f"Protein{i}" for i in range(12)]
```

Samples group by *why* the model scores them; the class sidebars show the two groups separate:

```python
import matplotlib.pyplot as plt
aa.plot_settings()
aa.AAPredPlot().clustermap(data=data, names=names, labels=labels)
plt.show()
```

**Further parameters.** ``colors`` maps each class to a sidebar color, ``cmap`` sets the heatmap colormap, ``figsize`` sizes the figure, and ``cbar_label`` / ``title`` label the colorbar and figure:

```python
aa.AAPredPlot().clustermap(
    data=data, names=names, labels=labels,
    colors={1: "tab:red", 0: "tab:gray"}, cmap="RdBu_r",
    figsize=(8, 8), cbar_label="Pearson correlation (r)",
    title="Explanation-similarity clustering")
plt.show()
```
