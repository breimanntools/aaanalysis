``AAPred().predict_domain()`` scores a domain and scans its boundary, shifting ``tmd_start`` / ``tmd_stop`` by every offset in ``[-window, +window]``:

```python
import aaanalysis as aa
aa.options["verbose"] = False  # Disable verbosity

df_seq = aa.load_dataset(name="DOM_GSEC")
labels = df_seq["label"].to_list()
df_feat = aa.load_features(name="DOM_GSEC").head(20)

# Training feature matrix
sf = aa.SequenceFeature()
X = sf.feature_matrix(features=df_feat, df_parts=sf.get_df_parts(df_seq=df_seq))

# Bind df_feat so raw sequences are featurized internally by the predict_* methods
aapred = aa.AAPred(df_feat=df_feat, random_state=42)
aapred.fit(X, labels)
```

The result shows how the score depends on the exact boundary; ``is_best`` flags the highest-scoring definition (offset ``0`` is the annotated boundary):

```python
df_domain = aapred.predict_domain(df_seq[df_seq["entry"] == "P05067"], window=3)
aa.display_df(df_domain, n_rows=10, show_shape=True)
```
