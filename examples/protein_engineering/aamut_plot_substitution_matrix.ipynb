:meth:`AAMutPlot.substitution_matrix` shows the 20x20 mean-absolute substitution impact as a heatmap.

```python
import matplotlib.pyplot as plt
import aaanalysis as aa
aa.plot_settings()
df_impact = aa.AAMut().run()
aa.AAMutPlot().substitution_matrix(df_impact=df_impact)
plt.tight_layout()
plt.show()
```

```python
# cmap sets the heatmap colormap; figsize sets the figure size in inches
aa.plot_settings()
aa.AAMutPlot().substitution_matrix(df_impact=df_impact, cmap="magma", figsize=(8, 7))
plt.tight_layout()
plt.show()
```
