To demonstrate the ``TreeModel().select_features()`` method, we obtain the ``DOM_GSEC`` example dataset and its feature set (see [Breimann25a]_), create the feature matrix, and fit a ``TreeModel`` to obtain Monte Carlo feature importance:

```python
import aaanalysis as aa
aa.options["verbose"] = False # Disable verbosity

df_seq = aa.load_dataset(name="DOM_GSEC")
labels = df_seq["label"].to_list()
df_feat = aa.load_features(name="DOM_GSEC").head(50)

# Create feature matrix and fit the model
sf = aa.SequenceFeature()
df_parts = sf.get_df_parts(df_seq=df_seq)
X = sf.feature_matrix(features=df_feat["feature"], df_parts=df_parts)

tm = aa.TreeModel(random_state=0)
tm.fit(X, labels=labels)
```

The ``'top_k'`` strategy keeps the ``param`` features with the highest feature importance:

```python
df_top10 = tm.select_features(df_feat=df_feat, strategy="top_k", param=10)
aa.display_df(df_top10)
```

The ``'threshold'`` strategy keeps features whose importance is at least ``param``:

```python
df_thr = tm.select_features(df_feat=df_feat, strategy="threshold", param=1.0)
aa.display_df(df_thr)
```

The ``'frequency'`` strategy keeps features selected in at least ``param`` (a fraction) of the per-round masks. These masks only vary when recursive feature elimination is enabled, so fit with ``use_rfe=True`` first:

```python
tm_rfe = aa.TreeModel(random_state=0)
tm_rfe.fit(X, labels=labels, use_rfe=True, n_feat_min=10, n_feat_max=25)
df_freq = tm_rfe.select_features(df_feat=df_feat, strategy="frequency", param=0.5)
aa.display_df(df_freq)
```
