``SequenceFeatureTransformer`` runs leak-free CPP feature selection behind the scikit-learn ``fit`` / ``transform`` API. First, the ``DOM_GSEC`` dataset:

```python
import aaanalysis as aa
aa.options["verbose"] = False  # Disable verbosity

# DOM_GSEC example dataset (df_seq with an 'entry' + 'sequence' column, one row per protein)
df_seq = aa.load_dataset(name="DOM_GSEC", n=20)
labels = df_seq["label"].to_list()
```

Construct the transformer with the CPP selection levers: the split configuration ``split_kws`` and scales ``df_scales``, the number of features ``n_filter``, the ``label_test`` / ``label_ref`` groups, the ``max_overlap`` / ``max_cor`` redundancy caps, whether to ``simplify`` the selected set, and ``n_jobs`` / ``random_state`` / ``verbose``:

```python
sft = aa.SequenceFeatureTransformer(
    split_kws=aa.SequenceFeature.get_split_kws(split_types=["Segment", "Pattern"]),
    df_scales=aa.load_scales(name="scales"),
    n_filter=50, label_test=1, label_ref=0,
    max_overlap=0.5, max_cor=0.5, simplify=False,
    n_jobs=1, random_state=42, verbose=False)

X = sft.fit_transform(df_seq, labels)
aa.display_df(sft.df_feat_, n_rows=10, show_shape=True)
```
