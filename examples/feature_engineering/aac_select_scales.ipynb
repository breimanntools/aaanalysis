The ``AAclust().select_scales()`` method returns a redundancy-reduced subset of an amino acid scales DataFrame in a single call. It clusters the scales and keeps one representative scale (medoid) per cluster, returning the matching columns of the input ``df_scales`` directly — ready to hand to ``CPP``.

```python
import aaanalysis as aa
aa.options["verbose"] = False

# Load amino acid scales (rows = amino acids, columns = scale IDs)
df_scales = aa.load_scales()
aa.display_df(df_scales, n_rows=10, show_shape=True)
```

Provide ``df_scales`` and a pre-defined number of clusters via ``n_clusters`` to obtain exactly that many representative scales. The returned DataFrame keeps the amino acid index and only the selected scale columns:

```python
aac = aa.AAclust()
df_scales_selected = aac.select_scales(df_scales, n_clusters=100)
aa.display_df(df_scales_selected, n_rows=10, show_shape=True)
```

The fitted attributes remain accessible on the instance, e.g. the names of the selected (medoid) scales, which equal the returned columns:

```python
print("Selected scales:", df_scales_selected.shape[1])
print("Columns match medoid names:",
      list(df_scales_selected.columns) == aac.medoid_names_)
```

If ``n_clusters`` is omitted (``None``), AAclust optimizes the number of clusters automatically. We use a smaller subset here for speed:

```python
df_scales_small = df_scales[list(df_scales.columns)[:50]]
df_auto = aac.select_scales(df_scales_small)
print("Optimized number of selected scales:", df_auto.shape[1])
```

The clustering optimization is controlled by ``min_th`` (Pearson correlation threshold) and ``on_center`` (apply ``min_th`` to the cluster center vs. to all members):

```python
# Higher threshold on all members -> stricter, more clusters
df_all = aac.select_scales(df_scales_small, on_center=False, min_th=0.5)
print("min_cor_all, min_th=0.5:", df_all.shape[1])
# Same threshold on the center -> less strict, fewer clusters
df_center = aac.select_scales(df_scales_small, on_center=True, min_th=0.5)
print("min_cor_center, min_th=0.5:", df_center.shape[1])
```

The optional cluster-merging step uses the ``metric`` parameter (distance/similarity measure for merging and medoid selection):

```python
for metric in ["euclidean", "correlation", "manhattan", "cosine"]:
    df_m = aac.select_scales(df_scales, n_clusters=100, metric=metric)
    print(f"{metric:>12}: {df_m.shape[1]} scales")
```

The returned ``df_scales_selected`` is a valid scales DataFrame and can be passed straight to ``CPP`` to build features on a redundancy-reduced scale set:

```python
df_cat = aa.load_scales(name="scales_cat")
df_cat_selected = df_cat[df_cat["scale_id"].isin(df_scales_selected.columns)]
aa.display_df(df_cat_selected, n_rows=10, show_shape=True)
```
