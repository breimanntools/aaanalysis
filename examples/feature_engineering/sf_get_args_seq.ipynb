The ``SequenceFeature().get_args_seq()`` method extracts the JMD-N, TMD, and JMD-C sequence parts of a single protein and returns them as a ready-to-use ``args_seq`` dictionary. This is the per-protein sequence input for sample-level plots such as :meth:`CPPPlot.profile` and :meth:`CPPPlot.feature_map` (e.g. for SHAP-based explanations), passed via ``**args_seq``.

We first load the DOM_GSEC example dataset:

````python
import pandas as pd
import aaanalysis as aa
aa.options["verbose"] = False  # Disable verbosity

df_seq = aa.load_dataset(name="DOM_GSEC", n=3)
sf = aa.SequenceFeature()
aa.display_df(df_seq, n_rows=10, show_shape=True)
````

Select a protein by its ``entry`` (accession) via the ``sample`` parameter. The returned dictionary carries the ``jmd_n_seq``, ``tmd_seq``, and ``jmd_c_seq`` parts:

````python
args_seq = sf.get_args_seq(df_seq=df_seq, sample="P05067")  # APP
aa.display_df(pd.DataFrame([args_seq]), char_limit=50)
````

Alternatively, a protein can be selected by its row position in ``df_seq``:

````python
args_seq = sf.get_args_seq(df_seq=df_seq, sample=0)
aa.display_df(pd.DataFrame([args_seq]), char_limit=50)
````

The flanking JMD lengths are controlled by ``jmd_n_len`` and ``jmd_c_len`` (default 10 each):

````python
args_seq = sf.get_args_seq(df_seq=df_seq, sample="P05067", jmd_n_len=5, jmd_c_len=5)
aa.display_df(pd.DataFrame([args_seq]), char_limit=50)
````

The returned dictionary is designed to be splatted directly into the sample-level plotting methods, removing the manual ``get_df_parts`` slicing glue:

```python
cpp_plot.profile(df_feat=df_feat, col_imp="feat_impact_APP", shap_plot=True, **args_seq)
cpp_plot.feature_map(df_feat=df_feat, name_test="APP", shap_plot=True, **args_seq)
```
