``SequenceFeature().scale_composition()`` builds a **baseline feature set** for a prediction model. For each sequence it averages every scale over the residues of a span, giving the ``(n_seq, n_scales)`` matrix ``X`` — the sequence's mean profile in scale-space (the scale-based analogue of amino-acid composition), with no positional information. Its purpose is **comparison**: fit the same classifier on this ``X`` and on a :class:`CPP` ``feature_matrix``, and compare the scores to see how much CPP's positional Part-Split-Scale features add over a plain scale average. Here we load the ``DOM_GSEC`` example dataset and the bundled scales (see [Breimann25]_):

```python
import aaanalysis as aa
aa.options["verbose"] = False
df_seq = aa.load_dataset(name="DOM_GSEC")
df_scales = aa.load_scales()
sf = aa.SequenceFeature()
```

By default (``list_parts=None``) the whole TMD-JMD span (``jmd_n`` + ``tmd`` + ``jmd_c``) is averaged. Residues that are not in the scale index (gaps ``'-'`` and other non-canonical symbols) are dropped before averaging:

```python
X = sf.scale_composition(df_seq=df_seq, df_scales=df_scales)
print(f"n samples: {X.shape[0]}")
print(f"n scales:  {X.shape[1]}")
print(f"Shape of X: {X.shape}")
```

With ``return_df=True`` the matrix is returned as a labeled ``pd.DataFrame`` (rows indexed by protein entry, one column per scale):

```python
df_scale_composition = sf.scale_composition(df_seq=df_seq, df_scales=df_scales, return_df=True)
aa.display_df(df_scale_composition, n_rows=10, show_shape=True)
```

Use ``list_parts`` to average only over selected sequence parts (concatenated per sequence). For example, restrict the baseline to the TMD:

```python
X_tmd = sf.scale_composition(df_seq=df_seq, df_scales=df_scales, list_parts="tmd")
print(f"Shape of X (TMD only): {X_tmd.shape}")
```

**Further parameters.** ``df_scales`` defaults to the bundled scales (or ``options['df_scales']``) when omitted, so ``sf.scale_composition(df_seq=df_seq)`` returns the scale-average matrix over the default scale set.

```python
X_default = sf.scale_composition(df_seq=df_seq)
print(f"Shape of X (default scales): {X_default.shape}")
```
