``transform`` builds the numeric CPP feature matrix from the features selected in ``fit``, so it is leak-free inside a scikit-learn ``Pipeline``. First, the dataset:

```python
import aaanalysis as aa
aa.options["verbose"] = False  # Disable verbosity

# DOM_GSEC example dataset (df_seq with an 'entry' + 'sequence' column, one row per protein)
df_seq = aa.load_dataset(name="DOM_GSEC", n=20)
labels = df_seq["label"].to_list()
```

Fit, then transform a sequence DataFrame ``X`` into the feature matrix:

```python
import pandas as pd
sft = aa.SequenceFeatureTransformer(n_filter=50, random_state=42).fit(df_seq, labels)
X_mat = sft.transform(X=df_seq)
aa.display_df(pd.DataFrame(X_mat, columns=sft.features_), n_rows=10, show_shape=True)
```
