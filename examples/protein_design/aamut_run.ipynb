``AAMut`` quantifies how substituting one amino acid for another shifts each physicochemical scale, independent of any sequence or task. :meth:`AAMut.run` returns a tidy per-scale substitution-impact table; here we restrict to the ``M -> V`` / ``M -> A`` substitutions over a few scales.

```python
import aaanalysis as aa
aam = aa.AAMut()
df_impact = aam.run(from_aa="M", to_aa=["V", "A"])
df_impact.head()
```

Leaving ``from_aa`` / ``to_aa`` unset returns all 20x19 ordered pairs; ``scales`` restricts to a subset of scale ids.
